```python
import os
import pandas as pd
from datetime import datetime

def extract_date_from_file(filepath):
    for enc in ['utf-8', 'windows-1252', 'latin1']:
        try:
            with open(filepath, 'r', encoding=enc) as f:
                for line in f:
                    line_clean = line.strip().lower().replace(" ", "")
                    if line_clean.startswith('%date:'):
                        # حذف اول % Date: و برداشتن باقی خط
                        date_str = line.split(":", 1)[1].strip()
                        try:
                            date_obj = datetime.strptime(date_str, "%d-%b-%y %H:%M:%S")
                            return date_obj
                        except ValueError:
                            print(f"Date format error in {filepath}: {date_str}")
            break
        except UnicodeDecodeError:
            continue
    return None



def process_signals_and_save(folder_path, output_excel_path):
    records = []
    
    files = sorted(os.listdir(folder_path))
    for filename in files:
        if filename.endswith(".txt"):  
            filepath = os.path.join(folder_path, filename)
            
            date_obj = extract_date_from_file(filepath)
            if not date_obj:
                print(f"Date not found in {filename}")
                continue

            signal_obj = TS(filepath)
            signal_obj.signal = signal_obj.signal / 9.806
            signal_obj.fftransform()

            iso_value = calculate_score_iso(signal_obj.rms())
            rms_value = signal_obj.rms()
            hump_range = detect_hump_range_clusters(signal_obj, top_percent=15)
            hump_result = check_hump_with_iso_ratio(signal_obj, hump_range, iso_value)

            records.append({
                "File Name": filename,
                "Date": date_obj.strftime("%Y-%m-%d %H:%M:%S"),
                "ISO": iso_value,
                "RMS": rms_value,
                "Hump Range": str(hump_range),
                "Hump Detected": hump_result
            })

            print(f"Processed {filename}")

    df = pd.DataFrame(records)
    df = df.sort_values("Date")
    df.to_excel(output_excel_path, index=False)
    print(f"Saved summary to {output_excel_path}")
```

```python
import matplotlib.pyplot as plt

def plot_signals_with_hump_ranges(signal_records):
    for record in signal_records:
        signal_obj = record["Signal Object"]
        hump_range = record["Hump Range"]

        hump_center = (hump_range[0] + hump_range[1]) / 2

        iso_score = calculate_score_iso(signal_obj.rms())
        rms_val = signal_obj.rms()
        hump_range_detected = hump_range
        hump_check = check_hump_with_iso_ratio(
            signal_obj,
            hump_range,
            iso_score
        )

        plt.figure(figsize=(12, 5))
        plt.plot(signal_obj.fq, signal_obj.ft, label=f"{record['File Name']}")

        plt.axvspan(hump_range[0], hump_range[1], color='red', alpha=0.3, label="Hump Range")
        plt.axvline(hump_center, color='blue', linestyle='--', linewidth=1.5, label=f"Hump Center: {hump_center:.1f} Hz")

        plt.xlim(0, 8000)
        plt.xlabel("Frequency (Hz)")
        plt.ylabel("Amplitude")
        plt.title(f"{record['File Name']} - {record['Date'].strftime('%Y-%m-%d %H:%M:%S')}")

        plt.legend()
        plt.grid(True)

        text_str = f"ISO Score: {iso_score:.4f}\nRMS: {rms_val:.4f}\nHump Range: {hump_range_detected}\nHump Detected: {hump_check}"
        plt.gcf().text(0.01, 0.01, text_str, fontsize=9, verticalalignment='bottom', bbox=dict(facecolor='white', alpha=0.5))

        plt.tight_layout()
        plt.show()

```

```python
def analyze_hump_trend(signal_records):
    hump_signals = []
    for record in signal_records:
        if check_hump_with_iso_ratio(
            record["Signal Object"],
            record["Hump Range"],
            calculate_score_iso(record["Signal Object"].rms())
        ):
            center_hump = (record["Hump Range"][0] + record["Hump Range"][1]) / 2
            hump_signals.append({
                "Date": record["Date"],
                "Center Hump": center_hump,
                "File Name": record["File Name"]
            })

    hump_signals.sort(key=lambda x: x["Date"])

    for i in range(1, len(hump_signals)):
        prev = hump_signals[i - 1]
        curr = hump_signals[i]

        if curr["Center Hump"] < prev["Center Hump"]:
            print(f"⚠️ Hump frequency moved lower from {prev['Center Hump']:.1f} Hz to {curr['Center Hump']:.1f} Hz between {prev['Date']} and {curr['Date']}")
        else:
            print(f"✔️ Hump frequency stable or higher: {prev['Center Hump']:.1f} Hz → {curr['Center Hump']:.1f} Hz")
```

```python
import numpy as np

def load_signal(filepath):
    with open(filepath, 'r') as f:
        lines = f.readlines()
    
    # حذف خطوط متادیتا
    data_lines = [line for line in lines if not line.startswith('%')]
    
    # تبدیل به آرایه عددی
    signal_values = np.array([float(line.strip()) for line in data_lines])
    
    # ساخت آبجکت Signal
    return TS(signal_values)
```

```python
import os
from datetime import datetime

signal_records = []

for file in os.listdir("../data/06FN04-Opal Pelletizing/06FN04-740 RPM-BSF Failure & BPFO-NU1048 NDE-NU1048+6048 DE/signals"):
    if file.endswith(".txt"):
        signal_obj = load_signal(os.path.join("../data/06FN04-Opal Pelletizing/06FN04-740 RPM-BSF Failure & BPFO-NU1048 NDE-NU1048+6048 DE/signals", file))

        date_str = file.split("_")[0]  # مثلا "2024-11-23_filename.txt"
        date = datetime.strptime(date_str, "%Y-%m-%d")

        hump_range = detect_hump_range_clusters(signal_obj, top_percent=15)

        signal_records.append({
            "Signal Object": signal_obj,
            "Date": date,
            "Hump Range": hump_range,
            "File Name": file
        })

analyze_hump_trend(signal_records)
```
